```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn.
# Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 
# Note:  timepoint, signal, event, and region are columns in the fmri dataset 




import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

plt.show()
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. 
# Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'. 


# Note:  pclass, age, and fare are columns in the titanic dataset. 

import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")



# Box plot for 'age'
plt.subplot(1, 2, 1)  # Create subplot 1
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)  # Create subplot 2
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()  # Optional: Adjust spacing between subplots

# Display the plots
plt.show()

```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. 
# Use the hue parameter for the 'cut' column of the diamonds dataset. 


import seaborn as sns
import matplotlib.pyplot as plt


diamonds = sns.load_dataset("diamonds")


sns.histplot(data=diamonds, x='price', hue='cut')


plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")


plt.show()

```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.


import seaborn as sns

iris = sns.load_dataset("iris")


sns.pairplot(data=iris, hue="species")

plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap

import seaborn as sns
import matplotlib.pyplot as plt


flights = sns.load_dataset("flights")
print(flights.head())


flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")


sns.heatmap(flights_pivot)

# Set plot title and labels
plt.title("Passenger Count Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
